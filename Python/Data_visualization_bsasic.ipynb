```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_survival_distribution_sns(data, column='survived', title='Survival Distribution'): #here you can write anything
   
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(data=data, x=column)

    # Add labels on each bar
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height}',(p.get_x() + p.get_width() / 2, height * 0.5), 
            ha='center', va='center', color='white', fontsize=10, fontweight='bold')


    ax.set_title(title)
    ax.set_xlabel(column.capitalize())
    ax.set_ylabel('Count')
    plt.tight_layout()
    plt.show()
```

#creating different differnt histogram plot for all numerical data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_all_numeric_histograms(data, columns=None, bins=10, color='skyblue'):
    """
    Plots histograms with internal data labels for all or selected numeric columns.

    Parameters:
    - data: DataFrame
    - columns: list of column names or None to auto-detect all numeric columns
    - bins: number of bins in histogram
    - color: color of bars
    """
    # If no columns specified, use all numeric columns
    if columns is None:
        columns = data.select_dtypes(include='number').columns.tolist()

    for column in columns:
        plt.figure(figsize=(6, 4))
        ax = sns.histplot(data[column].dropna(), bins=bins, color=color)

        # Add labels inside bars
        for patch in ax.patches:
            height = patch.get_height()
            if height > 0:
                ax.annotate(f'{int(height)}',
                            (patch.get_x() + patch.get_width() / 2, height * 0.5),
                            ha='center', va='center', color='white', fontsize=9, fontweight='bold')

        ax.set_title(f'Distribution of {column.capitalize()}')
        ax.set_xlabel(column.capitalize())
        ax.set_ylabel('Count')
        plt.tight_layout()
        plt.show()
```

 What is a Surface Plot?
A surface plot is a 3D visualization where:

X-axis = one variable

Y-axis = another variable

Z-axis = a value that depends on X and Y

👉 It’s like drawing a 3D “landscape” of your data.
It helps you see relationships between 3 variables at once.

💡 Real-world example: E-commerce
Imagine you want to analyze how Revenue depends on:

Price of product (X-axis)

Quantity purchased (Y-axis)

Revenue (Z-axis = Price × Quantity)

This is a perfect case for a surface plot, because revenue is determined by two inputs (price, quantity).



```python
import numpy as np
import plotly.graph_objects as go

# Example ranges for price ($) and quantity sold
price = np.linspace(10, 200, 30)     # from $10 to $200
quantity = np.linspace(1, 20, 30)    # from 1 to 20 units

# Create 2D grid of price and quantity
P, Q = np.meshgrid(price, quantity)

# Revenue = Price × Quantity
Revenue = P * Q

# Create surface plot
fig = go.Figure(data=[go.Surface(z=Revenue, x=P, y=Q, colorscale="Viridis")])

# Update layout
fig.update_layout(
    title="E-commerce Revenue Surface",
    scene=dict(
        xaxis_title="Price ($)",
        yaxis_title="Quantity",
        zaxis_title="Revenue ($)"
    ),width = 1000,
    height = 700
)

fig.show()
```

You own a pizza shop. You want to know:
How much money you make based on pizza price and number of pizzas sold



```python
import numpy as np
import pandas as pd

# Create sample data
np.random.seed(42)

data = pd.DataFrame({
    'Ad_Spend': np.random.randint(1000, 5000, 50),   # in $
    'Discount': np.random.randint(5, 30, 50)         # in %
})

# Sales depends on both
data['Sales'] = data['Ad_Spend'] * 0.5 + data['Discount'] * 100 + np.random.randint(-500, 500, 50)

print(data.head())
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['Ad_Spend'], data['Discount'], data['Sales'])

ax.set_xlabel('Ad Spend')
ax.set_ylabel('Discount')
ax.set_zlabel('Sales')

plt.title('3D Scatter Plot')
plt.show()
```

```python
from scipy.interpolate import griddata

# Create grid
x = data['Ad_Spend']
y = data['Discount']
z = data['Sales']

xi = np.linspace(x.min(), x.max(), 50)
yi = np.linspace(y.min(), y.max(), 50)
xi, yi = np.meshgrid(xi, yi)

zi = griddata((x, y), z, (xi, yi), method='cubic')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(xi, yi, zi)

ax.set_xlabel('Ad Spend')
ax.set_ylabel('Discount')
ax.set_zlabel('Sales')

plt.title('3D Surface Plot')
plt.show()

```

3. 3D Bar Plot
👉 Best for: Comparing grouped values

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = data['Ad_Spend'][:10]
y = data['Discount'][:10]
z = np.zeros(10)

dx = np.ones(10) * 200
dy = np.ones(10) * 2
dz = data['Sales'][:10]

ax.bar3d(x, y, z, dx, dy, dz)

ax.set_xlabel('Ad Spend')
ax.set_ylabel('Discount')
ax.set_zlabel('Sales')

plt.title('3D Bar Plot')
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Create range of values
P = np.linspace(10, 100, 30)   # Price
Q = np.linspace(1, 50, 30)     # Quantity

# Create grid (this is VERY IMPORTANT for surface plot)
P, Q = np.meshgrid(P, Q)

# Define Revenue function
Revenue = P * Q


'''
P = np.linspace(10, 100, 30)

10 (Start): The first number in your sequence.

100 (Stop): The last number in your sequence.

30 (Num): How many numbers you want in total.

The result: NumPy calculates the exact "step" size for you so that you get exactly 30 numbers starting at 10 and ending at 100.
'''
```

What this code means (VERY IMPORTANT)
🔹 np.meshgrid(P, Q)
Converts 1D arrays → 2D grid
Needed because:
Surface plot needs a grid of values
Not random points like scatter

👉 Think of it like a table:

Price	Quantity	Revenue
10	1	10
10	2	20
20	1	20
...	...	...
🔹 Revenue = P * Q
This is your mathematical model
For every (P, Q) pair → compute Revenue

👉 This creates the Z-axis values

```python
fig = go.Figure(data=[go.Surface(
            z=Revenue,   # height (Z-axis)
            x=P,         # X-axis (Price)
            y=Q,         # Y-axis (Quantity)
            colorscale="Viridis"
        )
    ]
)

fig.update_layout(
    title="Revenue Surface Plot",
    scene=dict(
        xaxis_title="Price",
        yaxis_title="Quantity",
        zaxis_title="Revenue"
    ),width = 1000,
    height = 700
)

fig.show()

```
