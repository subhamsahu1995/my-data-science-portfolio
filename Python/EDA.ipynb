# Measures of Dispersion / Second moment business decision
# The second moment in statistics refers to Variance and Standard Deviation, which are measures of dispersion — how spread out your data is around the mean.

```python
''' Insight example for understanding
Store	 Mean	Variance	Insight
Store A	 50.4	1.04	  Sales are consistent/stable
Store B	 52	    536	      Sales are highly inconsistent
So, even though both have similar average sales, the variance tells us that:

Store A is reliable and predictable.
Store B may need investigation due to unstable performance.
'''

'''
Store	Mean	Variance	Standard Deviation	Business Insight
Store A	50.4	1.04	1.02	Very stable sales
Store B	52	    536	    23.15	Highly variable and unpredictable
'''

education_data.var()
```

## Third moment business decision "Skewness"

## In simple words, it shows whether the data is tilted (skewed) more to the left or right.



## Let’s say you're analyzing customer spending, employee salaries, or delivery times.

If the data is not symmetric, your average (mean) can give a wrong impression.

Skewness helps you decide whether to use Mean or Median when summarizing data.
## positively skewed (Right-Skewed)
## Mean > Median > Mode

## Business impact?	Choose the right central value to act on
If skew is:

## ** Too high (>1 or <-1) → mean is not reliable → use median **

```python
## The 4th moment in statistics refers to kurtosis, which measures the "tailedness" of a distribution — in simple terms, how likely extreme values (outliers) are to occur.

print(education_data.workex.kurt())
print(education_data.gmat.kurt())
```

```python
'''
1. workex kurtosis = 13.40
Very high positive kurtosis
This is leptokurtic → sharply peaked with heavy tails
⚠️ It means:
Most candidates have similar work experience, but a few have very high or very low experience (extreme values).
Outliers are present
'''


'''
Kurtosis Value	Type	Shape	Outliers
≈ 0	Mesokurtic	Normal peak	Normal
> 0 and < 3	Slightly Leptokurtic	Slightly peaked	Few mild outliers
≫ 3 (e.g. > 10)	Very Leptokurtic	Very sharp peak and heavy tails	Many/extreme outliers
< 0	Platykurtic	Flat, wide distribution	Light tails
'''
```
